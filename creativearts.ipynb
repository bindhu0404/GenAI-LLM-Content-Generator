```python
from transformers import pipeline, AutoProcessor, MusicgenForConditionalGeneration
import torchaudio
import textwrap
```

```python
gen_pipe = pipeline("text-generation", model="gpt2")
```

```python
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")
```

```python
def generate_prose(prompt, max_length=200):
    output = gen_pipe(prompt, max_length=max_length, num_return_sequences=1)
    text = output[0]['generated_text']
    return "\n".join(textwrap.wrap(text, width=80))
```

```python
def generate_poetry(prompt, max_length=150):
    output = gen_pipe(prompt + "\n", max_length=max_length, num_return_sequences=1)
    text = output[0]['generated_text']
    return "\n".join(textwrap.wrap(text, width=60))
```

```python
def generate_music(prompt):
    inputs = processor(text=[prompt], return_tensors="pt")
    audio_values = model.generate(**inputs, max_new_tokens=256)
    torchaudio.save("generated_music.wav", audio_values[0].cpu(), 16000)
    return "generated_music.wav"
```

```python
type_of_content = input("What do you want to generate? (prose/poetry/music): ").strip().lower()
myprompt = input("Enter your prompt: ")
```

```python
if type_of_content == "prose":
    max_length = int(input("Enter maximum word limit (default is 200): ") or 200)
    prose = generate_prose(myprompt, max_length)
    print("\nGenerated Prose:\n", prose)
elif type_of_content == "poetry":
    max_length = int(input("Enter maximum word limit (default is 150): ") or 150)
    poetry = generate_poetry(myprompt, max_length)
    print("\nGenerated Poetry:\n", poetry)
elif type_of_content == "music":
    music_file = generate_music(myprompt)
    print("\nGenerated Music File:", music_file)
else:
    print("Invalid choice! Please enter 'prose', 'poetry', or 'music'.")
```
